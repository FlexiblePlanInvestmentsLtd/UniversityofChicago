1. **ROC (Rate of Change):** ROC compares the current price (via returns) to a past price (via compounded returns over the window). Positive values indicate upward momentum (gains), while negative values indicate downward momentum (losses). We calculated it as follows: $$ROC_t = \prod_{i=1}^n (1+r_i)-1$$

- For the weekly data in the broad assets, we would like to use 4, 6, 12, 26, 52-week ROC to capture momentum across short-term and long-term horizons.
- For the monthly data in the equity domestic / global assets, we would like to use 1, 3, 6, 9, 12-month ROC to capture momentum.

<span style="color:red">Note that ROC is the only indicator which will use the weekly / monthly returns data (other indicators all require to use the daily returns), we would calculate the ROC indicator outside the `apply_indicator` function by running the for loop to get the ROC for each asset. </span>

2. **MACD Line, Signal Line, and MACD Histogram:**
$$
\text{MACD Line} = \text{EMA}_{short} - \text{EMA}_{long}
$$

$$
\text{Signal Line} = \text{EMA}_{signal\,period}
$$

$$
\text{MACD Histogram} = \text{MACD Line} - \text{Signal Line}
$$
In this case, we tried the general one, in which we set the short-period as 12 days, long-period as 26 days, and signal period as 9 days.
- The MACD Line: This accounts for the change in momentum through the relation of short-term versus long-term trends. A positive MACD represents upward momentum, while a negative MACD represents downward momentum.
- Crossovers between the MACD Line and Signal Line act as buy/sell signals:
    - Bullish Crossover: The point at which the MACD Line crosses over the Signal Line.
    - Bearish Crossover: When the MACD Line crosses below the Signal Line.
- The MACD Histogram: focuses on the magnitude and direction of changes in momentum. An expanding histogram indicates strengthening momentum, while a shrinking histogram signals weakening momentum.

3. **RSI (Relative Strength Index) & Stochastic RSI:** RSI measures the relative strength of gains compared to losses, which indicates overbought or oversold conditions in an asset. Stochastic RSI enhances RSI by normalizing its values within a fixed range (0–1) and provides clearer signals in volatile markets.
First, we convert daily returns to cumulative price series:
$$
\text{price data}_t = \prod_{i=1}^n (1+r_i)
$$
Then we calculated the daily price changes $\Delta_t$. For positive price changes, we set negative changes to 0; For neagtive price changes, we absolute the values and set positive changes to 0. Compute the rolling average of gains and losses over the specified window period.
$$
\text{gain}_t = \text{loss}_t = \Delta_t = \text{price data}_t - \text{price data}_{t-1}
$$

$$
\text{average gain}_t = \frac{1}{\text{window}}\sum_{i=0}^{\text{window}-1}\text{gain}_{t-i}
$$

$$
\text{average loss}_t = \frac{1}{\text{window}}\sum_{i=0}^{\text{window}-1}\text{loss}_{t-i}
$$
The relative strength is then calculated to measure the ratio of average gains to average losses.
$$
\text{RS}_t = \frac{\text{average gain}_t}{\text{average loss}_t}
$$
The RSI is computed from RS using the following formula:
$$
\text{RSI}_t = 100 - \frac{100}{1+\text{RS}_t}
$$
- RSI values range from 0 to 100.
- $\text{RSI} > 70$: Asset is considered overbought (potential sell signal).
- $\text{RSI} < 30$: Asset is considered oversold (potential buy signal).

The Stochastic RSI can also be calculated by normalizing RSI values between 0 and 1 by comparing the RSI to its historical minimum and maximum over the rolling window:
$$
\text{Stoch RSI}_t = \frac{\text{RSI}_t - min(\text{RSI}_{t-window:t})}{max(\text{RSI}_{t-window:t}) - min(\text{RSI}_{t-window:t})}
$$
- $\text{Stoch RSI} > 0.8$: Overbought conditions.
- $\text{Stoch RSI} < 0.2$: Oversold conditions.

We leveraged 14-day (half a month) window to evaluate the RSI for each asset.

4. **Bollinger Bands (SMA, Upper, and Lower Bands):** Bollinger Bands can be effectively used as momentum indicators as we may observe a strong upward momentum when prices repeatedly touch or exceed the upper band, and vice versa. In another aspect, we can attain some of the volatility insights from it. In other words, a narrower band would always indicates lower volatility, which often preceds significant price movements.

In this implementation, we would opt 20, 90, 180, and 240-day as our rolling window sizes to perform the following bands calculations. For the rolling window size $N$:

We can calculate the simple moving average (SMA):
$$
\text{SMA}_N = \frac{1}{N}\sum_{i = t-N+1}^t P_i
$$

With SMA, it's easy for us to calculate the rolling standard deviation (SD):

$$
\text{SD}_N = \sqrt{\frac{1}{N}\sum_{i = t-N+1}^t (P_i - \text{SMA}_N)^2}
$$

It's easy for us to derive the Upper and Lower Bollinger Bands:
$$
\text{Upper}_N = \text{SMA}_N + \text{num\_std} \times \text{SD}_N
$$

$$
\text{Lower}_N = \text{SMA}_N - \text{num\_std} \times \text{SD}_N
$$

Where:
- $P_i$ is the price at time $i$, which has been converted by the returns data through `.cumprod()`.
- $N$ is the rolling window size.
- $\text{num\_std}$ is the number of standard deviations used (commonly 2).

5. **Hilbert Transform (DCPERIOD, DCPHASE, PHASOR, SINE, TRENDMODE)**: We also leveraged the convenience provided by `TA-Lib.HT` method to extract information about the instantaneous amplitude, phase, and frequency of a signal.

- Dominant Cycle Period (`HT_DCPERIOD`): Identifies the dominant cycle length in the price series to gauge the periodicity of market behavior.
- Dominant Cycle Phase (`HT_DCPHASE`): Measures the phase of the dominant cycle to provide insights into the cycle’s current position (e.g., peak, trough).
- Phasor Components (`HT_PHASOR`): Outputs in-phase and quadrature components to track the real-time cyclic structure of the time series.
- SineWave (`HT_SINE`): Generates sine and lead sine components to identify potential inflection points in market cycles.
- Trend vs Cycle Mode (`HT_TRENDMODE`): Distinguishes between trending and cyclic market conditions.

***

6. **EWMA (Exponentially Weighted Moving Average) Volatility:** We would like to substitute the traditional volatilities with the EWMA volatility by giving more weight to recent data while gradually discounting older observations with a decay factor (smoothing parameter, $\theta$), which would be more useful in rapidly changing markets. Besides filtering assets by momentum factors, we also want to guarantee the volatility of selecting assets considerable. On the other hands, changes in EWMA volatility can also signal potential momentum breakouts: Sudden increases in volatility often precede significant price movements. Stable, low EWMA volatility can indicate consolidation phases, while high volatility reflects turbulent periods.

In this case, we would set our initial variance as $\sigma_0^2 = (\frac{0.2}{\sqrt{252}})^2$. Then, we are able to evaluate the next-phase variance (and thus the volatility) using the previous-determined variance, $\sigma_{t-1}^2$ and the current-phase return $r_t$:

$$
\text{EWMA Volatility} = \sqrt{\sigma_t^2} = \sqrt{\theta \sigma_{t-1}^2 + (1-\theta)r_t^2}
$$

***

7. **Rolling Volatility / Rolling & Expanding Skew / Rolling & Expanding Kurtosis:** 
- High rolling volatility may indicate increased asset uncertainty; stable one suggests stable market conditions.
- Persistent negative rolling skewness suggests a greater likelihood of extreme negative returns (which would capture downside risk that momentums do not capture).
- High rolling kurtosis in conjunction with rising volatility may indicate potential asset instability.

The combination of these indicators would provide us a clearer regime identification for better asset allocation advise.
For rolling, we provided 20, 60, 126, 252, and 504-day windows to calculate metrics for comprehensive monitoring of short vs. long-term changes.

***

8. **Downside Deviation**: We also want to include this feature to quantify the standard deviation of negative asset returns relative to a specified target return (In default, we would like to set it as 0% and can be set to something positive if we really want to be more stringent). For each rolling window size in rolling_windows, the function calculates the downside deviation as:
$$
\text{Downside Deviation} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} \min(x_i - \text{Target Return}, 0)^2}
$$
Unlike standard deviation, which considers all variability, downside deviation isolates downside risk, thereby aligning better with risk-averse investors' perspectives.

***

9. **Cumulative Drawdown Periods:** The drawdown period indicator measures the number of consecutive periods an asset remains in a drawdown state. A drawdown occurs when the asset's cumulative return falls below its previous peak value. This indicator provides insights into how long an asset stays in a drawdown, which can be a good indicator to unveil the persistence of unfavorable market conditions. Here, we compute cumulative returns for each asset and identify the cumulative maximum returns (peak values) up to each time point; Determine if the current return is below the peak value (in drawdown) and count the number of consecutive periods the asset remains in drawdown until a new peak is achieved.

***

10. **Drawdown Counts:** The drawdown count indicator tracks the cumulative number of distinct drawdowns an asset has experienced up to each time point. A distinct drawdown begins when the asset enters a drawdown state after reaching a new peak. Different from cumulative one, drawdown counts firstly computes cumulative returns and identify peak values up to each time point and detect transitions from non-drawdown (0) to drawdown (1); Increment the count for each new drawdown initiation and then aggregate the counts over time for a cumulative total.

***

11. **Lagged Indicators:** We also want to include the lagged momentum indicators — such as rolling volatility, skewness, kurtosis, or other measures with lags of 1, 2, 4, 6, and 12 periods to capture the evolution of momentum over time, if any. Lagged indicators allow for detecting the acceleration or deceleration of trends that traditional momentum cannot have. If momentum weakens at longer lags, it could signal a potential reversal. Conversely, consistent strength across lags may confirm a robust trend. We created a multiple lag periods to provide a temporal cross-validation for momentum signals to ensure that the trend is not isolated to a specific time frame.

## Asset-wise Feature DataFrame Output for Equity Domestic / Global Baskets

To handle multiple N/A values for Equity Domestic & Global Baskets daily return data, we adjusted the function from `apply_indicators_broad_assets` to `apply_indicators_equity_baskets` by taking out the one-hot encoding part. Instead of directly calling the function, we would create an outer loop to drop the N/A values dynamically for each asset based on the length of its N/A value-period. In addition, we added the one-hot encoding asset columns to ensure the accurate implementation by adding the outer loop to iterate each asset. For reference, there are 46 output excel files for equity domestic assets and 52 output excel files for equity global assets in the `~/data` directory, named with `equity_domestic_n_features.xlsx` and `equity_global_n_features.xlsx`, respectively, where 'n' stands for the corresponding asset number for a specific basket.

```python
roc_periods = [1, 3, 6, 9, 12]
lags = [1, 2, 4, 6, 12]
asset_columns = [col for col in weekly_rets.columns if col.startswith('Asset')]

indicators = {
    'ROC': lambda df: calculate_roc(df, roc_periods)
}

asset_dataframes_master_domestic = {}

for i_investment in range(equity_domestic_monthly_rets.shape[1]):
    data_in = pd.DataFrame(equity_domestic.iloc[:, i_investment]).dropna()
    equity_domestic_monthly_rets_in = pd.DataFrame(equity_domestic_monthly_rets.loc[data_in.index[0]:])
    columns_in = data_in.columns

    downside_deviation = calculate_downside_deviation_ts(
        data=data_in, 
        rolling_windows=[20, 60, 126, 252, 504], 
        target_return=0
    )

    asset_dataframes_out = apply_indicators_equity_baskets(
        data=equity_domestic_monthly_rets_in,           # Weekly returns data for each asset
        indicators=indicators,                          # Dictionary of indicators
        assets=columns_in,                              # List of asset columns
        reference_data=data_in                          # Daily return data for indicators calculation
    )

    for asset, asset_df in asset_dataframes_out.items():
        # Add downside deviation for the current asset
        if not downside_deviation.empty:
            downside_dev_cols = [col for col in downside_deviation.columns if asset in col]
            if downside_dev_cols:  # Only add columns if they exist for the current asset
                asset_df = pd.concat([asset_df, downside_deviation[downside_dev_cols]], axis=1)

        for lag in lags:
            lagged_downside_dev = downside_deviation[downside_dev_cols].shift(lag).add_suffix(f"_lag{lag}")
            asset_df = pd.concat([asset_df, lagged_downside_dev], axis=1)

        one_hot_df = pd.DataFrame(0, index=asset_df.index, 
                                  columns=[f'Asset_{i+1}' for i in range(equity_domestic_monthly_rets.shape[1])])
        one_hot_df[f'Asset_{i_investment + 1}'] = 1  # Set 1 only for the current asset being processed
        asset_df = pd.concat([asset_df, one_hot_df], axis=1)
        asset_df = asset_df.dropna()
        asset_df.columns = [col.replace(f'_{asset}', '') for col in asset_df.columns]
        asset_dataframes_out[asset] = asset_df
    asset_dataframes_master_domestic[i_investment] = asset_dataframes_out

i_investment = 0
asset_name = 'Asset 1'

asset_dataframes = asset_dataframes_master_domestic[i_investment]
final_asset_df = asset_dataframes['Asset 1']
final_asset_df
```

```python
data_path = os.path.dirname(os.getcwd()) + '/data'
os.makedirs(data_path, exist_ok=True)

for i_investment, asset_dataframes_out in asset_dataframes_master_domestic.items():
    for asset, df in asset_dataframes_out.items():
        filename = os.path.join(data_path, f'equity_domestic_{i_investment + 1}_features.xlsx')
        df.to_excel(filename, sheet_name=f'{asset}_features', index=True)
        print(f"Exported {asset} to {filename}")
```
